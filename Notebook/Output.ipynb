```python
df = pd.read_excel(file_path1,sheet_name='Feuil1')
df.head()
```

```python
df.to_csv('csv_VP', index=False, sep=',', encoding='utf-8')
print("Succes")
```

```python
 
#Définir le seuil d'attrition
seuil_attrition = -90
#Créer les colonnes pour Encours Moyen par Année
em_pivot = df.pivot(index='ID client', columns='Année', values='Encours_Moyen_Cumulé').fillna(0) 
em_pivot.columns = [f'EM_{int(col)}' for col in em_pivot.columns]
 
#Créer les colonnes pour nombre de mois avec EM > 0 par an
app_pivot = df.pivot(index='ID client', columns='Année', values='Apparition_EM_Par_Mois').fillna(0) 
app_pivot.columns = [f'MoisActifs_{int(col)}' for col in app_pivot.columns]
 
#Calcul des évolutions EM
years = sorted(df['Année'].unique()) 
evolution = pd.DataFrame(index=em_pivot.index) 
for i in range(1, len(years)): 
    prev_year = years[i-1] 
    curr_year = years[i] 
    evol_col = f'Evolution_EM_{curr_year}vs{prev_year}'
    evolution[evol_col] = ( (em_pivot[f'EM_{curr_year}'] - em_pivot[f'EM_{prev_year}']) / em_pivot[f'EM_{prev_year}'].replace(0, 1) ) * 100
    #Forcer les valeurs > 100% a 101%
    evolution[evol_col] = evolution[evol_col].apply(lambda x : 101 if x > 100 else x)
#Ajouter la colonne d'attrition basée sur le seuil
def detect_attrition(row): 
    for col in row.index: 
        if col.startswith('Evolution_EM_2023vs2022') and row[col] <= seuil_attrition: 
            return 1
    return 0
#Finaliser
final_df = pd.concat([em_pivot, app_pivot, evolution], axis=1).reset_index()
#Appliquer attrition 
final_df['Attrition'] = final_df.apply(detect_attrition, axis=1)
#Afficher
final_df.head()
 
```

```python
# Export final
final_df.to_excel("clients_attrition2023.xlsx", index=False)
print("Fichier exporté avec succès.")
```
